```python
import pandas as pd
df2 = pd.read_csv('WorldCupMatches.csv')
df2['Year'] = df2['Year'].fillna(0).astype(int)
df2.head(3)
```

```python
df2 = df2.rename(columns={'Year':'Ano', 'Datetime':'Data Hora', 'Stage':'Grupo','Stadium':'Estádio', 'City':'Cidade', 'Home Team Name':'Equipe da Casa','Home Team Goals':'Gols Equipe Casa', 'Away Team Goals':'GolsVisitante','Away Team Name':'Equipe Visitante','Attendance':'Publico','Referee':'Juiz'})
df2.head(2)
```

```python
# Maior publico em cada grupo de todas as copas.
df22 = df2.groupby(['Grupo']).apply(lambda df2: df2.loc[df2.Publico.idxmax()])
df22.head(5)
```

```python
df2.describe()
```

```python
# Média de publico por estádio
df2[['Publico','Estádio']].groupby(['Estádio']).mean()
```

```python
# Quais os 10 copas com maiores publicos
df2['Ano'] = df2['Ano'].astype(int)
df_f = df2.pivot_table(index=['Ano'],values=['Publico'],aggfunc='sum') 
df_f.sort_values(by='Publico', ascending=False).head(10)
```
